```python
goalkeeper = ['GK']
defender = ['RCB', 'CB', 'LCB', 'LB', 'RB', 'RWB', 'LWB']
midfielder = ['CM','RCM', 'LCM', 'LDM', 'RDM', 'CAM', 'CDM', 'RM', 'LM', 'RAM', 'LAM']
attacker = ['RF', 'LF', 'ST', 'LW', 'RW', 'RS', 'LS', 'CF']
```

```python
def eleven(file, number):
    df = pd.read_csv(file)
    goalie = df.loc[df['Position'].isin(goalkeeper)].sort_values(by="Overall", ascending=False)[:1]
    defe = df.loc[df['Position'].isin(defender)].sort_values(by="Overall", ascending=False)[:4]
    mid = df.loc[df['Position'].isin(midfielder)].sort_values(by="Overall", ascending=False)[:5]
    att = df.loc[df['Position'].isin(attacker)].sort_values(by="Overall", ascending=False)[:1]
    result = pd.concat([goalie,defe,mid,att])
    result.to_csv("C:/Users/olafp/Desktop/ml/projekt mecze/jedenastki/2019/jed"+str(number)+".csv", encoding='utf-8')
```

```python
for i in range(20):
    eleven("C:/Users/olafp/Desktop/ml/projekt mecze/teamy/2019/team"+str(i)+".csv", i)
```

```python
results = np.array([0]*len(data19))
```

```python
#0 -> draw
#1 -> home team wins
#2 -> away team wins
for i in range(len(data19)):
    result = 0
    if data19.iloc[i,3] > data19.iloc[i,4]:
        result = 1
    elif data19.iloc[i,3] == data19.iloc[i,4]:
        result = 0
    else:
        result = 2
    results[i] = result
```

```python
np.unique(results, return_counts=True)
```

```python
results_pd = pd.concat([data19['HomeTeam'], data19['AwayTeam'], pd.DataFrame(results)], axis=1).rename(columns={0:"Results"})
```

```python
results_pd.to_csv('results.csv', encoding='utf-8')
```
