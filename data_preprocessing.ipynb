```python
tab = []
for i in range(20):
    team_t = pd.read_csv("C:/Users/olafp/Desktop/ml/projekt mecze/jedenastki/2019/jed"+str(i)+".csv")
    tab.append(team_t['Overall'])
```

```python
tab = pd.concat(tab, axis=1)
```

```python
teams = ['Arsenal', 'Bournemouth', 'Brighton', 'Burnley', 'Cardiff', 
         'Chelsea', 'Crystal Palace', 'Everton', 'Fulham', 'Huddersfield',
         'Leicester', 'Liverpool', 'Man City', 'Man United', 'Newcastle',
         'Southampton', 'Tottenham', 'Watford', 'West Ham', 'Wolves']
```

```python
results = pd.read_csv("C:/Users/olafp/Desktop/ml/projekt mecze/results.csv")
```

```python
results
```

```python
def create_instance(team1, team2, X):
    team1f = tab.iloc[:,teams.index(team1)]
    team2f = tab.iloc[:,teams.index(team2)]
    features = pd.concat([team1f,team2f])
    X.append(list(features))
```

```python
X = []
Y = []
for i in range(len(results)):
    create_instance(results.iloc[i,1],results.iloc[i,2],X)
    Y.append(results.iloc[i,3])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=0)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel = "linear").fit(X_train, Y_train)
```

```python
svm_predictions = model.predict(X_test)
```

```python
accuracy = model.score(X_test, Y_test)
```

```python
accuracy
```

```python
def predict_teams(team1, team2):
    X = [list(pd.concat([tab.iloc[:,teams.index(team1)],tab.iloc[:,teams.index(team2)]]))]
    return model.predict(X)
```
